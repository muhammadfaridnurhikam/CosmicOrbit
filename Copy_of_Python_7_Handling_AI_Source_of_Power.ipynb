## Write Text File

Pada program sebelumnya, kita hanya bisa me-read text file. Untuk me-write  file text kita perlu menambahkan argument **mode** pada function `open()`. Berikut adalah mode yang bisa kita gunakan untuk me-write file:

<br>

Mode | Keterangan
-|-
`r+` | read dan meniban konten file paling depan dengan data baru. File yang dibuka harus ada dalam directory terlebih dahulu.
`w` | mode untuk **meniban** seluruh konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`a` | mode untuk menambah konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`w+` | sama seperti mode `w` tetapi juga bisa melakukan read
`a+` | sama seperti mode `a` tetapi juga melakukan read

<br>

Jika argument mode tidak di-passing, function `open()` akan secara otomatis memilih mode `r` (read-only).

Coba jalankan program berikut kemudian analisis hasilnya!

r+.PNG

**Note:** mode `r` dan `r+` hanya bisa membuka file yang sudah tersedia dalam directory.

```python
f = open('/content/sample.txt','r+')
f.write('belajar python')
f.seek(0)
print(f.read())
f.close()
```

Contoh operasi yang bisa kita lakukan terhadap pandas Series & DataFrame:

`apply()` digunakan untuk menerapkan function atau lambda pada semua data dalam Series atau DataFrame.


operasidf2 (1).png



Pahami program secara perlahan dan saksama!

```python
def tentukan_grade(nilai):
  if nilai >= 80:
    return 'A'
  elif nilai >= 70:
    return 'B'
  elif nilai >= 60:
    return 'C'
  elif nilai >= 50:
    return 'D'
  else:
    return 'E'

nilai_akhir = data_nilai['Absensi']*0.3 + data_nilai['Ujian']*0.4
print(nilai_akhir)

grade = nilai_akhir.apply(tentukan_grade)
print(grade)

status = grade.apply(lambda x:'Lulus' if x!='c' else 'tidak lulus')
print(status)

data_nilai['nilai_akhir'] = nilai_akhir
data_nilai['grade'] = grade
data_nilai['status'] = status
data_nilai
```

Sebuah image tersusun dari banyak elemen kecil yang disebut sebagai pixel. Bagi kita, tiap pixel merepresentasikan warna tertentu. Tetapi, bagi komputer, tiap pixel berupa nilai dari warna tertentu.

Kita melihat pixel sebagai kumpulan warna, sedangakan komputer melihat pixel sebagai kumpulan nilai/angka. Nilai pixel umumnya disimpan dalam tumpukan matriks (tensor).

Kita bisa melihat tensor `lenna.png` dengan mencetak `img_lenna`.

printobject.PNG

```python
print(img_lenna)
```

Kita telah membuka image dengan function `open()` sebelumnya dan mendapat hasil berupa kumpulan binary. Sedangkan, ketika kita membuka image dengan bantuan OpenCV, kita malah mendapatkan pixel. Lalu, apa hubungan antara binary dan pixel?

Hubungannya ialah binary merupakan bentuk lain dari pixel. Seperti yang kita tahu, komputer hanya bisa memproses binary, sementara pixel bisa bernilai dari 0 sampai dengan 255 (bukan binary). Supaya bisa memproses pixel, komputer perlu mengkonversi pixel menjadi binary terlebih dahulu.

## Extract

Terdapat beberepa cara yang bisa kita gunakan untuk mengekstrak file compressed, diantaranya:

<br>

Jenis File Compressed | Syntax Extract
-|-
Extract file zip | `!unzip [lokasi file/nama file] -d [lokasi folder output]`
Extract file rar | `!unrar x [lokasi file/nama file] [lokasi folder output]`
Extract file tar | `!tar -xvf [lokasi file/nama file] -C [lokasi folder output]`
Extract file tar.gz | `!tar -xzvf [lokasi file/nama file] -C [lokasi folder output]`
Extract file tar.bz2 | `!tar -xjvf [lokasi file/nama file] -C [lokasi folder output]`

<br>

**Note:** `[lokasi folder output]` bisa diabaikan.
